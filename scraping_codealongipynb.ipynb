```

import time
def get_soup(url):
  return BeautifulSoup(requests.get(url).content, features="lxml")


def get_datafeatures(soup):
  return soup.title

def get_new_link(soup):
  return 'https://www.youtube.com' + soup.find_all('a', {"class": " content-link spf-link yt-uix-sessionlink spf-link "})[0]['href']


def _main(seed_url):
  mylist = []
  i = 0
  soup = get_soup(seed_url)
  mylist.append(get_datafeatures(soup))
  
  while i < 10:
    try:
      url = get_new_link(soup)
      soup = get_soup(url)
      mylist.append(get_datafeatures(soup))
      i +=1
      print(mylist)
    except:
      time.sleep(1)
      print(soup.content)
      print('got exception - restart')
      _main(seed_url)


  print(mylist)


_main('https://www.youtube.com/watch?v=CdhqVtpR2ts')
```

```
# this part of the code paralelizes the function 

# concept is simple, you want to perform 10 times the same function. Normally python will pick one processor from your computer and execute everything there
# if your computer has 2 processors, the process of paralelization is simple telling python to take the 10 tasks, split them between each processor, and each does 5 each

# in high performance computing you then buy clusters of thousands of processors

# show high performance computing in physics -> http://epp.tecnico.ulisboa.pt/

import multiprocessing
pool = multiprocessing.Pool(2)

num_cores = multiprocessing.cpu_count()
num_cores

pool.map(_main,['https://www.youtube.com/watch?v=CdhqVtpR2ts','https://www.youtube.com/watch?v=VDvr08sCPOc'])
```
