```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## How do I make a plot?

1 - I can plot functions (list of points) 

2 - I can plot multiple functions on the same plot

```python
# Creating data using NumPy
x1 = np.arange(0,10,1)
x2 = np.linspace(0,5,10)


plt.plot(x1, x1, label='linear')
#plt.plot(x1, x2**2, label='quadratic')   ## TODO: show the multiple plots on the same plot!

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```

## How do I make multiple plots?

```python
# Notice the use of np randn
np.random.seed(29)
data = np.random.randn(2, 100)
print(data[:56])
```

```python
#Histogram rapresents the frequency, bars rapresents the value

fig, axs = plt.subplots(2, 2, figsize=(20, 10))
axs[0, 0].hist(data[0])
axs[1, 0].scatter(data[0], data[1])
axs[0, 1].bar(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])

plt.show()
```

## How do I make a Pie Chart?

```python
labels = ['Vegetarians', 'Non-Vegetarians', 'Vegans']
sizes = [40, 45, 15]
```

## Getting Values from a dataset

```python
vehicles = pd.read_csv('vehicles.csv')

vehicles.sort_values(by=["CO2 Emission Grams/Mile", "City MPG", "Highway MPG"], inplace=True)
vehicles.plot(x="CO2 Emission Grams/Mile", y=["City MPG", "Highway MPG"])
```

```python
vehicles.head()
```

```python
vehicles_mean = vehicles[["Highway MPG", "City MPG", "Drivetrain"]].groupby(["Drivetrain"]).agg("mean")
vehicles_mean.plot.bar()
```

```python
vehicles_mean.plot.barh()
```

```python
plt.scatter(x = "Highway MPG", y = "City MPG", s = "Combined MPG", c = 'CO2 Emission Grams/Mile', data = vehicles, alpha = 0.5)
plt.show()
```

```python
vehicles.boxplot(column="Combined MPG")
```
