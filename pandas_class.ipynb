```
# how to perform binning -> bins on a variable

orders['Sales']
```

```
# this is the process of creating bins of a variable -> small boxes where you fit different ales values
"""
0 - 200 -> "low"
200 - 500 -> "medium"
501 - 1000 -> "high"
> 1000 -> "very high"
"""
```

```
# cut and 

labels = ["Very Low", "Low", "Moderate", "High", "Very High"]

# perform our bins

bins = pd.cut(orders['Sales'],5, labels = labels)

bins.value_counts()

```

```
# qcut

qbins = pd.qcut(orders['Sales'],5, labels = labels)

qbins.value_counts()
```

```
# a new column that subtracts the profit from the sales
orders['Cost'] = orders['Sales'] - orders['Profit']

# in pandas when you create new columns, operations are performed row by row
```

```
# Groupings -> Group by -> groupby(the dimensions you want to group by)[measures you want to aggregate].aggregation_method

orders.groupby(by = ['Segment'])['Sales'].sum()
```

```
# what if I want more than one aggregation method? -> write multiple aggregation methods
result = orders.groupby(by = ['Segment'])['Sales'].agg(["sum","mean","max"])
result
```

```
results = orders.groupby(by = ['Segment','Region'])['Sales'].agg(["sum","mean","max"])
```

```
# lets check out the index of this grouped by object

results.index
```

```
# in this case we have one level of group by (segment) but two metrics being aggregated
orders.groupby(by = ['Segment'])['Sales','Profit'].sum()
```

```
# what if I want to define my own aggregation operation?

# you define your own aggregation function

def range_aggregation(x):
  return max(x) - min(x)

range_aggregation([2,4,6,8,10]) 
```

```
# perform your usual group by, but then in the aggregation method, call the function you just defined
orders.groupby(by = ['Segment'])['Sales'].agg(range_aggregation)
```

```
# I want to learn how to apply better functions, row by row

orders['profit ratio'] = orders['Profit']/orders['Sales']
orders.head()
```

```
# what if I want to perform more complicated operations on the rows?

# I want to calculate if a row is profitable -> profit > 0?

# define a function that creates your logic

# the input will be the value of the profit of each row

def profitable(x):
  if x>=0:
    return 'Profitable'
  else:
    return 'Non-Profitable'

profitable(-10)
```

```
orders['Profit'].apply(profitable)
```

```
orders['Profit-Flag'] = orders['Profit'].apply(profitable)
orders.head()

# the apply method, will apply the function given, to each row in the dataframe, with the respective argument
```

```
# I want to create a new column that has the year in which the order was made
# first always test the logic of your function, on dummy data, NOT on the dataframe

def get_year(x):
  return str(x)[:4]

get_year('2017-02-07')
```

```
orders['year'] = orders['Order Date'].apply(get_year)
```

```
# alternative solution for dealing with the time stamp
# astpe function -> converts the entire data type of the column, to what you ask -> int, float, str
orders['Order Date'].astype(str).apply(get_year)
```
